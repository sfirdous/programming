```python
#String and String Methods
message = "hello, pyThon programming"
name = "Eda Lovelace"

print(message)
print(len(message))
print(message.upper())
print(name.lower())
print(message.title())
print(message.replace('pyThon','C++'))
```

```python
# string slicing 
print(name[0])
print(name[-1])
print(name[0:3])
print(name[-3:])
print(name[::2])
```

```python
# more string methods
print(message.count('hello'))
print(message.find('pr'))            # returns starting index of first occurence
print(message.split())               # returns array of words from sentence
first_name = name.split()[0]
print(first_name.split())           # does not splits word to letters just add it to the array of size 1
print(message.startswith('h'))      # character and word both work
print(message.endswith('programming'))
print('Python'.isalnum())           # true if all characters are letters or numbers and the string isnot empty
```
