```python
input_string = input('Enter a string')
vowels = 'aeiou'
input_string = input_string.lower()
vowels_count = 0
consonents_count = 0
for i in range(len(input_string)):
    if input_string[i] in vowels:
        vowels_count+= 1
    elif input_string[i].isalpha():
        consonents_count  += 1

print(input_string)
print(vowels_count,consonents_count)
```

```python
import numpy as np

numbers = np.random.randint(0,100,10)
list = [num for num in numbers]

sum = 0
avg = 0
len = 0

for l in list:
    sum += l
    len += 1
    
avg = sum/len
print(list)
print(sum,avg)
```

```python
largest = float('-inf')

for l in list:
    if l > largest:
        second_largest = largest
        largest = l
    elif l > second_largest and l != largest:
        second_largest = l
        
        
print(largest)
print(second_largest)
```

```python
unique = []

for l in list:
    if l not in unique:
        unique.append(l)
        
print(list)
print(unique)
```

```python
student_grades = {
    "Alice" : 50,
    "Bob" : 80,
    "Charlie" : 79,
    "David" : 90,
    "Eva" : 67,
    "Fred" : 88,
    "George" : 44,
    "Harry" : 30,
    "Iva" : 89,
    "John" : 64
}

    
marks = [student_grades.get(student,0) for student in student_grades]
maximum  = max(marks)

for key,val in student_grades.items():
    if val == maximum:
        print(f'{key} has the maximum marks')
        
print(sum(marks)/len(marks))

```

```python
reverse_dict = { marks:name for name,marks in student_grades.items()}
print(reverse_dict)
```
